```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the CSV file
df = pd.read_csv('C:/Users/charl/Documents/tesis/Bases/beeple_eth_trans.csv')
```

```python
# Step 2: Process the data
# Assuming the CSV has a 'timestamp' column and we want to count transactions per day
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date
daily_transactions = df.groupby('date').size()
```

```python
# Step 3: Plot the data
plt.figure(figsize=(15, 10))
plt.plot(daily_transactions.index, daily_transactions.values, marker='o', linestyle='-')
plt.xlabel('Fecha')
plt.ylabel('Numero de transacciones')
plt.title('Transacciones en ethereum')
plt.grid(True)
plt.xticks(rotation=45)
# Step 4: Highlight specific dates
# Step 4: Highlight specific dates
# Step 4: Highlight specific dates as separate data points
highlight_dates = ['2021-01-07', '2019-07-28', '2020-05-08', '2021-03-15', '2022-10-22', '2021-05-07']
highlight_dates = pd.to_datetime(highlight_dates).date

# Create a secondary y-axis for the highlighted dates
highlight_y = [daily_transactions.max() + 10] * len(highlight_dates)  # Position them above the main data

plt.scatter(highlight_dates, highlight_y, color='red', s=200, edgecolor='black', zorder=5)
for date in highlight_dates:
    plt.annotate(f'{date}',
                 xy=(date, daily_transactions.max() + 10),
                 xytext=(date, daily_transactions.max() + 15),
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="none"))
plt.tight_layout()
plt.show()
```
